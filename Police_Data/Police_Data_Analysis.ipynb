Police Data

```python
import pandas as pd
```

```python
df = pd.read_csv(r'PoliceData.csv');
```

```python
#showing the first N rows in the data (by default, N=5)

df.head(5)
```

```python
#detecting the missing values from each column of the dataframe.

df.isnull().sum()
```

```python
#To drop a column from dataframe.

df.drop(['country_name'], axis = 1)
```

```python
#In a column, show all the unique values with their count.

df['driver_age'].value_counts()
```

```python
#df.groupby(‘Col_1’)[‘Col_2’] .sum( ) - To create groups - Two Keys – Apply on Col_2 grouped by Col_1.

df.groupby('driver_gender').search_conducted.sum()
```

```python
df.stop_duration.value_counts()
```

```python
#df['Column_name'].map( { old1:new1 , old2:new2} ) – Change the all values of a column from old to new. We have to write for all values of column otherwise Nan will appear.

df['stop_duration']= df['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 24, '30+ Min' : 45})
```

```python
df
```

```python
#df['Column_name'].mean() - To show Mean value of a column.

df['stop_duration'].mean()
```

```python
#df.groupby('Column_1').Column_2.describe() - To create groups based on Column1 and show statistics summary based on Column2.

df.groupby('stop_date').driver_gender.describe()
```
